```python
# calculate WER for each sample file
from evaluate import load
import os

model_outputs_folder = "/content/drive/My Drive/research/all_transcripts/google speech transcripts/"
ground_truth_outputs_folder = "/content/drive/My Drive/research/all_transcripts/ground truth/"

filenames = os.listdir("/content/drive/My Drive/research/all_transcripts/ground truth/")

filenames.sort()

avg_score = 0
for filename in filenames:
  with open(model_outputs_folder+filename, 'r') as f:
    model_output = f.readlines()

  with open(ground_truth_outputs_folder+filename, 'r') as f:
    correct_output = f.readlines()
  
  wer = load("wer")
  wer_score = wer.compute(predictions=model_output, references=correct_output)
  print(filename)
  print(f"correct output: {correct_output}")
  print(f"model output: {model_output}")
  print(f"WER: {wer_score}\n\n")
  avg_score += wer_score


avg_score /= len(filenames)
print("Average WER Score: ", avg_score)
```

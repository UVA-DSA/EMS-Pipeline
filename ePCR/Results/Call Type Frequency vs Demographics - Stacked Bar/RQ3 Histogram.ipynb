```python
'''
    RQ3: Relationship between weight, age, gender and Chief Complaint/Call TypeÂ 
    1. Separate male & female
    2. Make sep dictionaries for each age group
    3. count frequency
'''

#importing what we need
import pandas as pd 
import numpy as np
```

```python
data = pd.read_excel('RQ3 Converted.xlsx').set_index('ID')

data.head(2)
```

```python
male_df = data[data['Gender'] == 'M']
male_df['Age'] = male_df.Age.astype(float)

female_df =  data[data['Gender'] == 'F']
female_df['Age'] = female_df.Age.astype(float)

male_df.dtypes
```

```python
#make function to return age group

def get_ag(age):
    if 0 <= age and age <3:
        return 0
    elif 3 <= age and age <6:
        return 1
    elif 6 <= age and age <14:
        return 2
    elif 14 <= age and age <19:
        return 3
    elif 19 <= age and age <34:
        return 4
    elif 34 <= age and age <49:
        return 5
    elif 49 <= age and age <65:
        return 6
    elif 65 <= age and age <79:
        return 7
    elif 79 <= age:
        return 8

#make function to add to dictionary
def add_to_dict(c, dictionary):
    if not pd.isnull(c):
        if c not in dictionary: 
            dictionary[c] = 1
        else:
            dictionary[c] += 1
```

```python
m_ct = [{}, {}, {}, {}, {}, {}, {}, {}, {}]
m_cc = [{}, {}, {}, {}, {}, {}, {}, {}, {}]


for index, row in male_df.iterrows():
    age = row['Age']
    ccs = row['Uniform Chief Complaints'][1:-1].split('}{')
    ct =  row['CallType']
    if not pd.isna(age):
        #print(age, get_ag(age))
        #add ccs and ct to dictionaries
        add_to_dict(ct, m_ct[get_ag(age)])
        for cc in ccs:
            add_to_dict(cc, m_cc[get_ag(age)])
        #print(m_ct[get_ag(age)])

f_ct = [{}, {}, {}, {}, {}, {}, {}, {}, {}]
f_cc = [{}, {}, {}, {}, {}, {}, {}, {}, {}]

for index, row in female_df.iterrows():
    age = row['Age']
    ccs = row['Uniform Chief Complaints'][1:-1].split('}{')
    ct =  row['CallType']
    if not pd.isna(age):
        #add ccs and ct to dictionaries
        add_to_dict(ct, f_ct[get_ag(age)])
        for cc in ccs:
            add_to_dict(cc, f_cc[get_ag(age)])

```

```python
f_ct
```

```python
# convert dictionary to dataframes

m_ct_df = pd.DataFrame.from_dict(m_ct[0], orient='index', columns=[0])
m_cc_df = pd.DataFrame.from_dict(m_cc[0], orient='index', columns=[0])
f_ct_df = pd.DataFrame.from_dict(f_ct[0], orient='index', columns=[0])
f_cc_df = pd.DataFrame.from_dict(f_cc[0], orient='index', columns=[0])

for i in range(1, len(m_ct)):
    m_ct_df = pd.concat([m_ct_df, pd.DataFrame.from_dict(m_ct[i], orient='index', columns=[i])], axis=1).fillna(0)
    m_cc_df = pd.concat([m_cc_df, pd.DataFrame.from_dict(m_cc[i], orient='index', columns=[i])], axis=1).fillna(0)
    f_ct_df = pd.concat([f_ct_df, pd.DataFrame.from_dict(f_ct[i], orient='index', columns=[i])], axis=1).fillna(0)
    f_cc_df = pd.concat([f_cc_df, pd.DataFrame.from_dict(f_cc[i], orient='index', columns=[i])], axis=1).fillna(0)
```

```python

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('RQ3 Histogram Data.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
m_ct_df.to_excel(writer, sheet_name='Male Call Types')
m_cc_df.to_excel(writer, sheet_name='Male Chief Complaints')
f_ct_df.to_excel(writer, sheet_name='Female Call Types')
f_cc_df.to_excel(writer, sheet_name='Female Chief Complaints')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python

#importing what we need
import pandas as pd 

data = pd.read_excel('RAA Data with Uniform CT & A,G,W & Uniform CC.xlsx')

data.head(2)
```

```python
#define methods

#make function to add to dictionary
def add_to_dict(c, dictionary):
    if not pd.isnull(c):
        if c not in dictionary: 
            dictionary[c] = 1
        else:
            dictionary[c] += 1
```

```python
ct_freq = {}
cc_freq = {}

for index, row in data.iterrows():
    ccs = row['Uniform Chief Complaints'][1:-1].split('}{')
    ct =  row['CallType']
    
    add_to_dict(ct, ct_freq)
    for cc in ccs:
        add_to_dict(cc, cc_freq)
```

```python
ct_df = pd.DataFrame.from_dict(ct_freq, orient='index')
cc_df = pd.DataFrame.from_dict(cc_freq, orient='index')
```

```python
ct_df
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Frequency of CTs and CCs.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
ct_df.to_excel(writer, sheet_name='Call Types')
cc_df.to_excel(writer, sheet_name='Chief Complaints')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

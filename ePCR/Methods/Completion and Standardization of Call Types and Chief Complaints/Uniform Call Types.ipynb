```python
import pandas as pd
```

```python
data = pd.read_excel('RAA Data with Age, Gender, and Weight & Uniform CC.xlsx')

data.head(2)
```

```python
#call types dictionary

dict_df = pd.read_excel('Dictionary for Call Types.xlsx').fillna('')
temp_dict = dict_df.set_index('Call Type').T.to_dict('list')

ct_dict = {}
for call_type, variations in temp_dict.items():
    for var in variations:
        if var != '':
            ct_dict[var] = call_type

```

```python
ct_dict
```

```python
new_cts = []

for index, row in data.iterrows():
    current_ct = row['CallType']
    new_ct = ct_dict[current_ct]
    new_cts.append(new_ct)
    #print(current_ct, row['CallType'])
```

```python
data['CallType'] = new_cts

data
```

```python
data.to_excel('RAA Data with Uniform CT & A,G,W & Uniform CC.xlsx')
```

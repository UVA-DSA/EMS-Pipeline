```python
'''
Mapping Chief Complaints from "key words"
    0. Read in RAA Data with separated cc section
    1. Read in Complied List of CC
    2. Convert Complied List of CC to dictionary
        {cheif complaint: [Key words columns]}
    3. Make new column for "uniform" chief complaints as an array 
    4. For each row in RAA Data, iterate through keys and see which chief complaints it matches
    NOTE: for more information, in case ChiefComplaint is null, combine extracted chief complaint column 
    with original chiefcomplaint column
        For each row,
            new_row = ""
            for cc in dict.keys
                for regex_string in dict[cc]
                includes_cc = row[CheifComplaint].to_lowercase().includes(dict[regex_string])
                if includes_cc: 
                    new_row += {cc}
                    break
            #after you go through all cc in dict
            uniform_cc_arr.append(new_row)
    5. If chief complaint does not map to a specif CC, see which general category it falls into

'''

import pandas as pd
```

```python
data = pd.read_excel('FINAL Separated Narratives.xlsx').set_index('ID')
```

```python
data
```

```python
dict_df = pd.read_excel('Complied List of Chief Complaints.xlsx', sheet_name = 'Regex Mapping for Specific CC')
```

```python
dict_df.fillna('', inplace=True)
dict_df.drop(dict_df.columns.difference(['Chief Complaint','Key words 1', 'Key words 2']), 1, inplace=True)
dict_df
```

```python
#convert to dictionary
dictionary = dict_df.set_index('Chief Complaint').T.to_dict('list')
dictionary
```

```python
#do the same for other dictionary

other_dictionary_df = pd.read_excel('Complied List of Chief Complaints.xlsx', sheet_name = 'Regex Mapping for Other')
other_dictionary_df.fillna('', inplace=True)
other_dict = other_dictionary_df.set_index('Chief Complaint').T.to_dict('list')
other_dict
```

```python
uniform_cc_arr = []
import re

# For each row,
#             new_row = ""
#             for cc in dict.keys
#                 for regex_string in dict[cc]
#                 includes_cc = row[CheifComplaint].includes(dict[regex_string])
#                 if includes_cc: 
#                     new_row += {cc}
#                     break
#             #after you go through all cc in dict
#             uniform_cc_arr.append(new_row)

cc_combined = data['ChiefComplaint'].fillna('') + data['Chief Complaint'].fillna('')

for row in cc_combined:
    new_row = ''
    #print(row)
    try: #if there is a chief complaint
        for cc in dictionary.keys():
            for regex_string in dictionary[cc]:
                if regex_string != '':
                    #print(regex_string)
                    includes_cc = re.search(regex_string, row.lower())
                    if includes_cc: 
                        #print(regex_string, cc)
                        new_row += '{' + cc + '}'
                        break
        #after you go through all cc in dict
        if new_row == '': #if there was no regex mapping
            for cc in other_dict.keys(): #try to map to a more general cheif complaint
                for regex_string in other_dict[cc]: 
                    if regex_string != '':
                        #print(regex_string)
                        includes_cc = re.search(regex_string, row.lower())
                        if includes_cc: 
                            new_row += '{' + cc + '}'
                            #print('CC:',new_row, 'narr:', row)
                            break
        #after you tried all cheif complaints in general dictionary, if there is STILL no match
        #print('new row:', new_row)
        if new_row == '': 
            #if cc_combined starts with . as first non-whitespace char, then cc is missing
            if row == '' or row.lstrip()[0] == '.':   
                new_row = '{Missing}'
            else:
                #print('new row:', new_row)
                #print('CC:','Other', row)
                new_row = '{Other}'  #label cc as 'Other' 
        #print(new_row, row)
        uniform_cc_arr.append(new_row)
    except AttributeError as e: #if there is no chief complaint
        uniform_cc_arr.append(new_row) #append blank string


#uniform_cc_arr
        
    
```

```python
new_data = data.copy()
new_data.insert(2, 'Uniform Chief Complaints', uniform_cc_arr)

new_data
```

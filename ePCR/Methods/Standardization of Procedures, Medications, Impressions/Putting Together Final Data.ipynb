```python
import pandas as pd

```

```python
# read in 2017-2018 files

df17rxcui = pd.read_excel('RAA Data 2017 RXCUI with ID cleaned.xlsx')
df17icd = pd.read_excel('RAA Data 2017 ICD-10-CM with ID cleaned.xlsx')
df17snomed =pd.read_excel('RAA Data 2017 Snomed with ID cleaned.xlsx')

df18rxcui = pd.read_excel('RAA Data 2018 RXCUI with ID cleaned.xlsx')
df18icd = pd.read_excel('RAA Data 2018 ICD-10-CM with ID cleaned.xlsx')
df18snomed =pd.read_excel('RAA Data 2018 Snomed with ID cleaned.xlsx')
```

```python
df17rxcui['Impression'] = df17icd['Impression']
df17rxcui['Procedures'] = df17snomed['Procedures']

df18rxcui['Impression'] = df18icd['Impression']
df18rxcui['Procedures'] = df18snomed['Procedures']
```

```python
df17rxcui
```

```python
#read in other files

df19 = pd.read_excel('RAA Data 2019 with ID cleaned.xlsx')
df20 = pd.read_excel('RAA Data 2020 with ID cleaned.xlsx')
```

```python
# write out all years separately

df20.to_excel('FINAL 2020 RAA Data.xlsx')
df19.to_excel('FINAL 2019 RAA Data.xlsx')
df18rxcui.to_excel('FINAL 2018 RAA Data.xlsx')
df17rxcui.to_excel('FINAL 2017 RAA Data.xlsx')
```

```python
# merge all data

dfFINAL = pd.concat([df20, df19, df18rxcui, df17rxcui])

dfFINAL.to_excel('COMBINED FINAL RAA Data, 2017-2020.xlsx')
```

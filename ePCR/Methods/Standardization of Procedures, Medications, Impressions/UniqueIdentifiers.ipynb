```python
#importing everything we need

import pandas as pd 
```

```python
# read in data

years = ["2020", "2019", "2018", "2017"] #if we get new data, we can just add that here
dataframes = []

for year in years: 
    dataframe = pd.read_excel('RAA Data ' + year + '.xlsx')
    dataframes.append(dataframe)

print(dataframes[1].iloc[8018]['Impression'])
```

```python
for i in range(len(years)): 
    idlist = [] #declare a list of unique identifiers that will be converted into a column
    df = dataframes[i]
    
    for j in range(2, df.shape[0] + 2):
        idlist.append(years[i] + '_' + str(j))
        
    df.insert(0, "ID", idlist, True) #insert ids into dataframes
    
    
```

```python
#check dataframes

for df in dataframes:
    print(df)
    
print(dataframes[1].iloc[5913]['Impression'])
```

```python
#write out new data 

for i in range(len(years)): 
    dataframes[i].to_excel('RAA Data ' + years[i] + ' with ID.xlsx')
```

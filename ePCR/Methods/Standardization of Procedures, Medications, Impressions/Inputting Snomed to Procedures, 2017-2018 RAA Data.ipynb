```python
'''
    import dictionaries
    write to files
    format {Medication/Prod/Impress (UniqueID)}
'''
import pandas as pd

years = ["2018", "2017"] #if we get new data, we can just add that here
dataframes = []


for year in years: 
    dataframe = pd.read_excel('RAA Data ' + year + ' separated with ID cleaned.xlsx')
    dataframes.append(dataframe)
```

```python
# MEDICATIONS Snomed 

#import pro dict
%store -r 
    
pro_id_dict = pd.read_excel('Complied list of Procedures.xlsx', sheet_name='Grouped Data' ).set_index('Snomed').T.to_dict('list')
pro_s_dict = {}
for rx, s_list in pro_id_dict.items():
    for i in range(len(s_list)):
        s = s_list[i]
        if s not in pro_s_dict:
            pro_s_dict[s] = ''
        pro_s_dict[s] = rx
        
#med dict
# key: string, procedure. value: INT, id
pro_s_dict
```

```python
#for each element in {}, add ID at end in (1234)

new_columns = []
missing_pros = set()

for i in range(len(years)):
    df = dataframes[i]
    
    col_wo_b = df['Procedures'].str[1:-1] #remove {} 
    procedures = col_wo_b.str.split('}{')
    
    col_w_id = []
    
    
    for row in range(len(procedures)):
        new_row = ''
        try: #for NaN values
            for each in procedures[row]:
                #get rid of timestamps. Find first instance of space
                start = each.find(' ') + 1
                
                time_stamp = each[0:start]
                each_cleaned = each[start:] 
                
                if each_cleaned in pro_s_dict:
                    snomed = pro_s_dict[each_cleaned]
                    new_row += '{' + time_stamp + each_cleaned + ' (' + str(snomed) + ')}' 

                else:
                    new_row += '{' + time_stamp + each_cleaned + ' (XXXX)}' 
                    missing_pros.add(each_cleaned)
            print(new_row)
            col_w_id.append(new_row)
                    
        except TypeError:
            col_w_id.append('') #blank string when there is no procedures
    
    #after complete df, add arrays to matrix
    new_columns.append(col_w_id)
```

```python
#replace procedures column with new columns

for i in range(len(years)):
    df = dataframes[i]
    new_col = new_columns[i]
    
    print(len(new_col))
    
    df['Procedures'] = new_col
```

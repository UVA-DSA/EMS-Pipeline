```python
'''
    RQ5: What are the most common sequences of treatments for the most common types of impressions?
    Pre-process to in order to make graph that represents "Markov Chain"
    Use time-series data -> make array of interventions ids separated by commas (NO SPACES) [432000,7806,2300]
        
'''

#importing what we need
import pandas as pd 
```

```python
data = pd.read_excel('RAA Data with Time-series.xlsx').set_index('ID')

data.head(2)
```

```python
df = data[['Impression', 'Time-series']]
df.head(2)
```

```python
#for top 4 impressions, convert time-series into array of intervention ids (EXCLUDING VITALS)
import re 

# Frequency of the medications/interventions prescribed for specific types of impressions, and their correspondings impressions
# corresponding impressions

#define all impressions to look at 
impressions_name = ['Seizures', 'Chest Pain (Angina/Cardiac)', 'Chest Pain (STEMI/MI)', 'Cardiac Arrest', 'Hypoglycemia', 'Stroke']
impressions_id = ['G40.90[91]', 'I20.[09]', 'I21.[4012]', 'I46.9', 'E(13.64)|E(16.2)', 'I6[32].9']

converted_arrays = []

for i in range(len(impressions_id)):
    #select all rows with that impression
    has_impression = data['Impression'].str.contains(impressions_id[i], regex=True, na=False)
    rows = data[has_impression]
    #print(rows)
    
    #initiliaze new column for this impression
    new = []
    
    for index, row in rows.iterrows():
        ts = row['Time-series'].replace("), (", ")(")
        #print(no_vitals_ts)
        #initalize array of ids
        id_arr = "start,"
        #get ids
        temp_arr = ts[3:-3].split(")(")
        #print(temp_arr)
        #print(index)
        for s in temp_arr: 
            t = tuple(s.replace("'","").split(', '))
            #filter - General - Contact Medical Control, Confirm Time, CV - ECG - 4 Lead Monitoring, 
            #CV - ECG - 12 Lead Obtained, IV Start - Extremity Vein (arm or leg), ECG monitoring
            filter_list = [
                 '304562007', '410669006' # , '428803005', '268400002', '392230005', '46825001',
            ]
            if t[0] != "Vital" and t[1] not in filter_list:
                id_arr += t[1] + ","
                

        #add id_arr to new column after you extract all ids for all rows
        id_arr += "end"
        #print(id_arr)
        new.append(id_arr)

    
    converted_arrays.append(new)
```

```python
#write out new columns

writer = pd.ExcelWriter( 'RQ4 WITH MOST COMMON - Graph Set 2 Data.xlsx', engine='xlsxwriter')

for i in range(len(impressions_name)):    
    ids = pd.DataFrame(converted_arrays[i], columns=['Sequence'])
    sn = impressions_name[i][0:31].replace('/', ' ')
    print(sn)
    ids.to_excel(writer, sheet_name=sn)
    
writer.save()
```

```python
import pandas as pd
```

```python
data = pd.read_excel('COMBINED FINAL RAA Data, 2017-2020.xlsx')

data.head(2)
```

```python
# MEDICATIONS

medications_tuples = []

medications_col = data['Medications']
medications_col = medications_col.str[1:-1] #remove {} 

medications_matrix = medications_col.str.split('}{')
    
for row in range(len(medications_matrix)):
    new_row = []
    try: #for NaN values
        for each in medications_matrix[row]:
            #get rid of timestamps. Find first instance of space
            start_time = each.find('T') + 1
            end_time = each.find(' ') - 1

            start_string = end_time + 2
            end_string = each.rfind('(') - 1

            start_med_id = end_string+2
            end_med_id = -1

            time_stamp = each[start_time:end_time] 
            string = each[start_string:end_string]
            med_id = each[start_med_id:end_med_id]

            new_row.append(('Medication', med_id, string, 'N/A', time_stamp))  
        #after complete row, add row to medication_tuples
        medications_tuples.append(new_row)

    except TypeError:
        medications_tuples.append(new_row)
        pass #skip and keep going if nan
```

```python
medications_tuples
```

```python
# PROCEDURES

procedures_tuples = []

procedures_col = data['Procedures']
procedures_col = procedures_col.str[1:-1] #remove {} 

procedures_matrix = procedures_col.str.split('}{')
    
for row in range(len(procedures_matrix)):
    new_row = []
    try: #for NaN values
        for each in procedures_matrix[row]:
            #get rid of timestamps. Find first instance of space
            start_time = each.find('T') + 1
            end_time = each.find(' ') - 1

            start_string = end_time + 2
            end_string = each.rfind('(') - 1

            start_pro_id = end_string+2
            end_pro_id = -1

            time_stamp = each[start_time:end_time] 
            string = each[start_string:end_string]
            pro_id = each[start_pro_id:end_pro_id]

            new_row.append(('Procedure', pro_id, string, 'N/A', time_stamp))
        #after complete row, add row to medication_tuples
        procedures_tuples.append(new_row)

    except TypeError:
        procedures_tuples.append(new_row)
        pass #skip and keep going if nan
```

```python
procedures_tuples
```

```python
# VITALS


    
    
vital_look_up = {
    'Pulse': 'Pulse',
    'Resp' : 'Respirations',
    'BP' : 'Blood Pressure',
    'GCS' : 'Glasgow Coma Scale',
    'Glucose' : 'Blood Sugar Level',
    'SPO2' : 'Pulse Oximetry',
    'Pain' : 'Pain Level',
    'EKG' : 'Electrocardiogram'   
}

vitals_tuples = []

vitals_col = data['Vitals']
vitals_col = vitals_col.str[1:-1] #remove {} 

vitals_matrix = vitals_col.str.split('}{')
    
for row in range(len(vitals_matrix)):
    new_row = []
    try: #for NaN values
        for each in vitals_matrix[row]:
            #get rid of timestamps. Find first instance of space
            start_time = each.find('T') + 1
            end_time = each.find(' ') - 1
            time_stamp = each[start_time:end_time] 

            vitals = each[end_time+2:].split(' ', 7)
            for vital in vitals:
                div = vital.find('-')
                string = vital[:div]
                value = vital[div+1:]
                if value == '': value = 'N/A'
                unified_concept = vital_look_up[string]
                new_row.append(('Vital', unified_concept, string, value, time_stamp))

              
        #after complete row, add row to medication_tuples
        vitals_tuples.append(new_row)

    except TypeError:
        vitals_tuples.append(new_row)
        pass #skip and keep going if nan
```

```python
vitals_tuples
```

```python
# JOIN ALL TUPLES TOGETHER AND SORT BY TIME

all_tuples = []

for i in range(len(medications_tuples)):
    joined = sorted(medications_tuples[i] + procedures_tuples[i] + vitals_tuples[i], key=lambda x: x[-1])
    all_tuples.append(joined)
```

```python
all_tuples
```

```python
data['Time-series'] = all_tuples

data
```

```python
data.to_excel('RAA DATA with Time-series.xlsx')
```

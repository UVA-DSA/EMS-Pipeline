```python
#importing what we need
import pandas as pd 

data = pd.read_excel('RAA Data with Uniform CC.xlsx')

data.head(2)
```

```python
#define regex strings

#for age
year_age_pattern = r'[0-9]{1,3}(?= ?y/o| ?year| ?yo| ?YO| ?Y/O)'
younger_age_pattern = r'[0-9]{1,3}(?: ?[Mn=m]onth(?!s|\.|,| ago)| ?minute(?!s|\.|,)| ?day(?!s|\.|,)| ?hour(?!s|\.|,))'

#for gender
female_pattern = r'(?<=yo)f|[Ff]emale| her| she| HER| SHE'
male_pattern = r'(?<=yo)m|(?<![Ff]e)[Mm]ale| his| (?<!s)he| HIS| (?<!S)HE'


#for weight
very_pattern = r'significantly (?=obese|overweight)|morbidly (?=obese|overweight)'
obese_pattern = r'obese|overweight'
weight_pattern = r'[0-9]+ ?(lbs|KG)'
```

```python
age_arr = []
gender_arr = []
weight_class_arr = []
weight_arr = []

narratives = + data['Narrative'].fillna('')  + data['Medic Notes'].fillna('') 

import re

count = 0

for narrative in narratives:
    try:
        weight = re.search(weight_pattern, narrative)
        weight_arr.append(weight.group(0)) if weight else weight_arr.append('')
        
        year_age = re.search(year_age_pattern, narrative)
        younger_age = re.search(younger_age_pattern, narrative)
        if year_age:
            age_arr.append(year_age.group(0) + ' year')
        elif younger_age:
            #print(younger_age.group(0))
            #sometimes, the narrative is incorrect and says patient that is adult's years and "minute/month/day/hour"
            #filter these out
            time_unit = younger_age.group(0)[younger_age.group(0).find(' '):]
            age = younger_age.group(0)[0:younger_age.group(0).find(' ')]
            if weight:
                age_arr.append(age + ' year') if int(weight.group(0)[0:weight.group(0).find(' ')]) > 12 else age_arr.append(younger_age.group(0)) 
                print(age + time_unit)
                count +=1
            else:
                if narrative.find('Pregnancy/Childbirth/Miscarriage') == -1 and (time_unit == ' minute' or time_unit == ' day' or time_unit == ' hour'):
                    #print(narrative)
                    age_arr.append(age + ' year')
                    print(age + time_unit)
                    count +=1
                elif (time_unit == ' month' or time_unit == 'Month') and int(age) > 13:
                    age_arr.append(age + ' year')
                    print(age + time_unit)
                    count +=1
                else:
                    age_arr.append(younger_age.group(0)) 
        else:
            age_arr.append('')
        
        if re.search(female_pattern, narrative):
            #print('F', '\n')
            gender_arr.append('F')
        elif re.search(male_pattern, narrative): 
            #print('M', '\n')
            gender_arr.append('M')
        else:
            #print('None', '\n')
            gender_arr.append('')
            
        weight_class = ''
        if re.search(very_pattern, narrative): weight_class += 'morbidly '
        if re.search(obese_pattern, narrative): weight_class += 'obese'
        weight_class_arr.append(weight_class)
        
        
        
        
    except TypeError as e: #if nan
        age_arr.append('')
        gender_arr.append('')
        weight_class_arr.append('')
        weight_arr.append('')


print(count)
```

```python
data['Age'] = age_arr
data['Gender'] = gender_arr
data['Weight'] = weight_arr
data['Weight Class'] = weight_class_arr

data
```

```python
data.to_excel('RAA Data with Age, Gender, and Weight & Uniform CC.xlsx')
```

```python
#importing everything we need

import pandas as pd 
```

```python
'''
Parsing:
Intervention/Procedures
    2017, 2018 parse by {}
    2019, 2020 prase by {
    
    2018 remove #) in the beginning of column showing the number of interventions taken
    2017,2018 PCR repetitive (means "patient care report")
    
    2017, 2018 - "Interventions"
    2019, 2020 - "Procedures" "Medications"
    
    Intervention: sub classes procedures and medications
    

Impression
    2017, 2018 parse by {}
    2019, 2020 parse by ), remove white space in front and quotation marks
    

To do:
    2018 remove #) in the beginning of column showing the number of interventions taken
    2019-2020 make interventions, procedures and medications parse by {}
'''
```

```python
#read in files

df2020 = pd.read_excel('RAA Data 2020 with ID.xlsx')
df2019 = pd.read_excel('RAA Data 2019 with ID.xlsx')
df2018 = pd.read_excel('RAA Data 2018 with ID.xlsx')
df2017 = pd.read_excel('RAA Data 2017 with ID.xlsx')
                       
```

```python
#remove numbers from 2018

df2018['Interventions'] = df2018['Interventions'].str[3:]
df2018 #check
```

```python
#remove PCR from 2017, 2018

df2018['Impression'] = df2018['Impression'].str.replace(' - PCR}', '}')
df2017['Impression'] = df2017['Impression'].str.replace(' - PCR}', '}')
```

```python
#add {} instead of "" in 2020 and 2019 for Cheif Complaint

df2020['ChiefComplaint'] = df2020['ChiefComplaint'].str[1:-1] #remove begin and end ""
df2019['ChiefComplaint'] = df2019['ChiefComplaint'].str[1:-1]

df2020['ChiefComplaint'] = df2020['ChiefComplaint'].str.replace('""', '}{') #remove begin and end ""
df2019['ChiefComplaint'] = df2019['ChiefComplaint'].str.replace('""', '}{')

df2020['ChiefComplaint'] = df2020['ChiefComplaint'].str.replace('"', '') #remove all other "
df2019['ChiefComplaint'] = df2019['ChiefComplaint'].str.replace('"', '')

df2020['ChiefComplaint'] = '{' + df2020['ChiefComplaint'] + '}' 
df2019['ChiefComplaint'] = '{' + df2019['ChiefComplaint'] + '}'

df2020
```

```python
#add {} instead of () and "" in 2020 and 2019

# replace all commas with whitespace
df2020['Impression'] = df2020['Impression'].str.replace(",", " ")
df2019['Impression'] = df2019['Impression'].str.replace(",", " ")

# remove all ""
df2020['Impression'] = df2020['Impression'].str.replace('"', '')
df2019['Impression'] = df2019['Impression'].str.replace('"', '')

# put { in the very begin. of the row amd } at end of row
df2020['Impression'] = '{'+ df2020['Impression'] + '}'
df2019['Impression'] = '{'+ df2019['Impression'] + '}'

# insert }{ after all ) following a number
df2020['Impression'] = df2020['Impression'].str.replace('(?<=[0-9|\.][0-9]|[0-9][A-Z])\)', ')}{', regex=True)
df2019['Impression'] = df2019['Impression'].str.replace('(?<=[0-9|\.][0-9]|[0-9][A-Z])\)', ')}{', regex=True)

#delete extra whitespace and end {}
df2020['Impression'] = df2020['Impression'].str.replace('{  ', '{').str.replace('{ ', '{')
df2019['Impression'] = df2019['Impression'].str.replace('{  ', '{').str.replace('{ ', '{')
df2020['Impression'] = df2020['Impression'].str.replace('{}', '')
df2019['Impression'] = df2019['Impression'].str.replace('{}', '') 

#check df2020[]
print(df2020['Impression'])
print(df2019['Impression'])
```

```python
# add {} instaed of { in 2019, 2020 in prodcedures and medications

for col in ['Procedures', 'Medications']:
    # add } after )
    df2020[col] = df2020[col].str.replace("{", "}{")
    df2019[col] = df2019[col].str.replace("{", "}{")

    df2020[col] = '{'+ df2020[col] + '}'
    df2019[col] = '{'+ df2019[col] + '}'

    df2020[col] = df2020[col].str.replace('{}', '')
    df2019[col] = df2019[col].str.replace('{}', '') 
```

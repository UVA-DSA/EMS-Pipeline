```python
'''
    import dictionaries
    write to files
    format {Medication/Prod/Impress (UniqueID)}
'''
import pandas as pd

years = ["2018", "2017"] #if we get new data, we can just add that here
dataframes = []


for year in years: 
    dataframe = pd.read_excel('RAA Data ' + year + ' separated with ID cleaned.xlsx')
    dataframes.append(dataframe)
```

```python
# MEDICATIONS RXCUI 


#med dict
# key: string, medicine. value: INT, id

temp = pd.read_excel('Complied list of ODEMSA Medications.xlsx', sheet_name='Grouped Data')
temp['RxCui'] = temp['RxCui'].astype(str)
med_dict =  temp.set_index('RxCui').T.to_dict('list')

med_s_dict = {}
for rxcui, array in med_dict
    med_s_dict[]
```

```python
#for each element in {}, add ID at end in (1234)

new_columns = []
missing_meds = set()

for i in range(len(years)):
    df = dataframes[i]
    
    col_wo_b = df['Medications'].str[1:-1] #remove {} 
    medications = col_wo_b.str.split('}{')
    
    col_w_id = []
    
    
    for row in range(len(medications)):
        new_row = ''
        try: #for NaN values
            for each in medications[row]:
                #get rid of timestamps. Find first instance of space
                start = each.find(' ') + 1
                
                time_stamp = each[0:start]
                each_cleaned = each[start:] 
                
                if each_cleaned in med_s_dict:
                    rxcui = med_s_dict[each_cleaned]
                    new_row += '{' + time_stamp + each_cleaned + ' (' + str(rxcui) + ')}' 
                else:
                    new_row += '{' + time_stamp + each_cleaned + ' (XXXX)}' 
                    missing_meds.add(each_cleaned)
            print(new_row)
            col_w_id.append(new_row)
                    
        except TypeError:
            col_w_id.append('') #blank string when there is no medications
    
    #after complete df, add arrays to matrix
    new_columns.append(col_w_id)
```

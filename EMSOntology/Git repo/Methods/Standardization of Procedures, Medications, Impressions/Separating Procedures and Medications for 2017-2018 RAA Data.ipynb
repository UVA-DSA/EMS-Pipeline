```python
#importing what we need
import pandas as pd 
```

```python

years = ["2018", "2017"] #if we get new data, we can just add that here
dataframes = []


for year in years: 
    dataframe = pd.read_excel('RAA Data ' + year + ' with ID cleaned.xlsx')
    dataframes.append(dataframe)
    
    #check for NaN values
    x = dataframe['Interventions'].isnull()
    nan_values = lambda x, xs: [i for (y, i) in zip(xs, range(len(xs))) if x == y]
    print(year + ' Null values in Interventions: ', nan_values(True, x)) #check for NaN values
```

```python
dataframes[0]
```

```python
pro_new_cols = []
med_new_cols = []
%store -r


    
med_id_dict = pd.read_excel('Complied list of ODEMSA Medications.xlsx', sheet_name='Grouped Data' ).set_index('RxCui').T.to_dict('list')
med_s_dict = {}
for rx, s_list in med_id_dict.items():
    for i in range(len(s_list)):
        s = s_list[i]
        if s not in med_s_dict:
            med_s_dict[s] = ''
        med_s_dict[s] = rx

        
for i in range(len(years)):
    df = dataframes[i]
    procedures = []
    medications = []
    
    df['Interventions'] = df['Interventions'].str[1:-1] #remove{}
    interventions = df['Interventions'].str.split('}{')
    
    for row in range(len(interventions)):
        row_procedures = ''
        row_medications = ''
        for each in interventions[row]:
            #get rid of timestamps. Find first instance of space
            start = each.find(' ') + 1
            time_stamp = each[0:start]
            each_cleaned = each[start:] 
            if each_cleaned in med_s_dict: #use rxcui string dict
                row_medications = row_medications + '{' + time_stamp + each_cleaned + '}'
            else: 
                row_procedures = row_procedures + '{' + time_stamp + each_cleaned + '}'
        procedures.append(row_procedures)
        medications.append(row_medications)
        
    pro_new_cols.append(procedures)
    med_new_cols.append(medications)
    
                
    
```

```python
#test 
print(len(pro_new_cols[0]) == len(med_new_cols[0]))
print(len(pro_new_cols[1]) == len(med_new_cols[1]))
print(med_new_cols[0])
print('\n')
print(pro_new_cols[0])
```

```python
for i in range(0,len(years)):
    df = dataframes[i]
    del df['Unnamed: 0']
    del df['Unnamed: 0.1']
    df.insert(loc=5, column='Procedures', value=pro_new_cols[i])
    df.insert(loc=6, column='Medications', value=med_new_cols[i])
    del df['Interventions']
    
    
```

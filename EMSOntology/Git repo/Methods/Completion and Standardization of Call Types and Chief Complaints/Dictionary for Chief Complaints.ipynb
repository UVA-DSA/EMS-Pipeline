```python
import pandas as pd
```

```python
years = ['2017','2018','2019','2020']

list_of_dataframes = []

for year in years:
    list_of_dataframes.append(pd.read_excel(f'RAA Data {year} with ID cleaned.xlsx'))

data = pd.concat(list_of_dataframes).set_index('ID')
```

```python
data
```

```python
#define dictionary of cief complaints

ungrouped_2017 = {}
ungrouped_2019 = {}

cc = data['ChiefComplaint'].str.split('}{')

for call_id, row in cc.iteritems():
    try:
        for c in row:
            complaint = c.replace('{', '').replace('}', '')
            
            #print(call_id[0:4])
            if call_id[0:4] == '2017' or call_id[0:5] == '2018':
                if complaint not in ungrouped_2017: ungrouped_2017[complaint] = 0
                ungrouped_2017[complaint] += 1
                if complaint == 'General Illness - V' or complaint == 'General Ill' or complaint == 'General Ill' or complaint == 'Genera' or complaint == 'General Illness': 
                    print(complaint, call_id)
            else:
                if complaint not in ungrouped_2019: ungrouped_2019[complaint] = 0
                ungrouped_2019[complaint] += 1
    except TypeError:
        pass #skip if nan

ungrouped_2017
```

```python
#convert dictionary to dataframe

ungrouped_df_2017 = pd.DataFrame(list(ungrouped_2017.values()), index=list(ungrouped_2017.keys())).reset_index()
ungrouped_df_2017 = ungrouped_df_2017.rename(columns={'index': 'CheifComplaint', 0: 'Frequency'})

ungrouped_df_2019 = pd.DataFrame(list(ungrouped_2019.values()), index=list(ungrouped_2019.keys())).reset_index()
ungrouped_df_2019 = ungrouped_df_2019.rename(columns={'index': 'CheifComplaint', 0: 'Frequency'})
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('Complied list of Chief Complaints.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
ungrouped_df_2017.to_excel(writer, sheet_name='2017-2018')
ungrouped_df_2019.to_excel(writer, sheet_name='2019-2020')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

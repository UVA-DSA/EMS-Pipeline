```python
'''
    subhead_set: {string "subheading"}
    
    Determine which dictionary each call belongs to by looking at the formatting of the first sub-section using regex expressions
    Make set of call_ids that do not fit into any of the three dictionaries above, in case of no regex match
        misfit_set = {string "call_id"}
'''

import pandas as pd #import what we need
import re  #regex
```

```python
#read in data
df = pd.read_excel('COMBINED FINAL RAA Data, 2017-2020.xlsx')

meddf = pd.read_excel('Complied list of ODEMSA Medications.xlsx')
```

```python
#make medications dictionary

meds = set()

for col in meddf.columns:
    meds_list = meddf[col].values.tolist()
    for each in meds_list: 
        meds.add(each)
meds
```

```python
df = df.set_index('ID')
```

```python
df
```

```python
#define regex string

letter_caps = r'((?<![a-zA-Z0-9\/\.])[A-Z]{1}(?::|-| -|\/ |\.))'
letter_lower = r'((?<![a-zA-Z0-9\/\.])[a-z]{1}(?:-|- )(?![0-9]))'
letter_caps_p = r'((?<![a-zA-Z0-9\/\.])\([A-Z]{1}(?:\)-|\)))'
word_lower = r'([A-Z][a-z]{2,}[\/ ]?(?:[A-Za-z]\s?\/?){0,}:)'
brackets = r'(\[.+?\])'
word_caps = r'(»?[A-Z]{2,}[ ]?(?:[A-Z]\s?\/?){0,}(?:-|:))'

regex_strings = [letter_caps, letter_lower, letter_caps_p, word_lower, brackets, word_caps]
regex_string_names = ['letter_caps', 'letter_lower', 'letter_caps_p', 'word_lower', 'brackets', 'word_caps']

print(len(regex_strings))
```

```python
#define dictionaries

# letter_caps_set = set()
# letter_lower_set = set()
# letter_caps_p_set = set()
# word_lower_set = set()
# brackets_set = set()
# word_caps_set = set()

# sets = [letter_caps_set, letter_lower_set, letter_caps_p_set, word_lower_set, brackets_set, word_caps_set]

tuple_set = set()
misfit_set = set()
nan_set = set()
```

```python
# #define method to extract regex match

# def extract(arrayoftuples):
#     array = []
#     for t in arrayoftuples:
#         for each in t:
#             array.append(each) if each != '' and len(each) != 1 else None
#     return array
```

```python
#for each narrative, search for all regex match and add matches to subhead_set. 
# If there are no regex matches, add call ID to misfit_set



for call_id, row in df.iterrows():
    try:
        narrative = row['Narrative']
        match_category = set()
        for i in range(len(regex_strings)):
            match_array = re.findall(regex_strings[i], narrative) #re.findall returns tuples based on groups
            #print(call_id,i, match_array)
            for each in match_array:
                if each[1:-1] not in meds: match_category.add(each)
        match_category = list(match_category)
        match_category.sort()
        #print(match_category)
        tuple_set.add(tuple(match_category)) if len(match_category) != 0 else misfit_set.add(call_id) 
    except TypeError as e: #nan values
        print(e)
        nan_set.add(call_id)
    
```
